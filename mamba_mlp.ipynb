```python
import os
import shutil
from torchvision.transforms import CenterCrop, Resize
from PIL import Image

def save_image_pairs(clean_image_paths, noisy_image_paths, train_dir, val_dir, split_ratio=1.0):
    # ëª¨ë“  clean/noisy ì´ë¯¸ì§€ ê²½ë¡œë¥¼ ìŒìœ¼ë¡œ ë¬¶ìŠµë‹ˆë‹¤.
    clean_to_noisy = {}
    for clean_path in clean_image_paths:
        clean_id = '_'.join(os.path.basename(clean_path).split('_')[:-1])
        clean_to_noisy[clean_id] = clean_path

    noisy_clean_pairs = []
    for noisy_path in noisy_image_paths:
        noisy_id = '_'.join(os.path.basename(noisy_path).split('_')[:-1])
        if noisy_id in clean_to_noisy:
            clean_path = clean_to_noisy[noisy_id]
            noisy_clean_pairs.append((noisy_path, clean_path))

    # ë°ì´í„° ë¶„í•  (Train/Validation)
    split_index = int(len(noisy_clean_pairs) * split_ratio)
    train_pairs = noisy_clean_pairs[:split_index]
    val_pairs = noisy_clean_pairs[split_index:]

    # ë””ë ‰í† ë¦¬ê°€ ì—†ìœ¼ë©´ ìƒì„±í•©ë‹ˆë‹¤.
    os.makedirs(train_dir, exist_ok=True)
    os.makedirs(val_dir, exist_ok=True)

    # ì´ë¯¸ì§€ ì „ì²˜ë¦¬ (Center Crop and Resize)
    center_crop = CenterCrop(1024)
    resize = Resize((64, 64))
    resize2 = Resize((64*4, 64*4))

    def save_pairs(pairs, target_dir):
        for noisy_path, clean_path in pairs:
            file_name = os.path.basename(clean_path)  # ë™ì¼í•œ íŒŒì¼ëª… ì‚¬ìš©
            noisy_save_path = os.path.join(train_dir, file_name)
            clean_save_path = os.path.join(val_dir, file_name)

            # ì´ë¯¸ì§€ë¥¼ ë¶ˆëŸ¬ì™€ ì „ì²˜ë¦¬ ì ìš© í›„ ì €ì¥
            noisy_image = Image.open(noisy_path).convert("RGB")
            clean_image = Image.open(clean_path).convert("RGB")

            # ì „ì²˜ë¦¬: ì¤‘ì•™ í¬ë¡­ í›„ ë¦¬ì‚¬ì´ì¦ˆ
            noisy_image = resize(center_crop(noisy_image))
            clean_image = resize2(center_crop(clean_image))

            noisy_image.save(noisy_save_path)
            clean_image.save(clean_save_path)

    # Trainê³¼ Validationì— ê°ê° ìŒì„ ì €ì¥í•©ë‹ˆë‹¤.
    save_pairs(train_pairs, train_dir)
    save_pairs(val_pairs, val_dir)

    print(f"{len(train_pairs)} training pairs and {len(val_pairs)} validation pairs saved.")

# ê²½ë¡œ ì„¤ì •
clean_image_folder = '/content/drive/MyDrive/hi/test/Input'
noisy_image_folder = '/content/drive/MyDrive/hi/test/Input'
train_folder = '/content/drive/MyDrive/hi/hi3'
val_folder = '/content/drive/MyDrive/hi/hi3'

clean_image_paths = [os.path.join(clean_image_folder, x) for x in os.listdir(clean_image_folder)]
noisy_image_paths = [os.path.join(noisy_image_folder, x) for x in os.listdir(noisy_image_folder)]

save_image_pairs(clean_image_paths, noisy_image_paths, train_folder, val_folder)
```

```python
import torch
import sys
sys.path.append('../content/NAFNet')
from basicsr.models import create_model
from basicsr.utils import img2tensor as _img2tensor, tensor2img, imwrite
from basicsr.utils.options import parse
import numpy as np
import cv2
import matplotlib.pyplot as plt

def imread(img_path):
  img = cv2.imread(img_path)
  img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
  return img
def img2tensor(img, bgr2rgb=False, float32=True):
    img = img.astype(np.float32) / 255.
    return _img2tensor(img, bgr2rgb=bgr2rgb, float32=float32)

def display(img1, img2):
  fig = plt.figure(figsize=(25, 10))
  ax1 = fig.add_subplot(1, 2, 1)
  plt.title('Input image', fontsize=16)
  ax1.axis('off')
  ax2 = fig.add_subplot(1, 2, 2)
  plt.title('NAFNet output', fontsize=16)
  ax2.axis('off')
  ax1.imshow(img1)
  ax2.imshow(img2)

def single_image_inference(model, img, save_path):
      model.feed_data(data={'lq': img.unsqueeze(dim=0)})

      if model.opt['val'].get('grids', False):
          model.grids()

      model.test()

      if model.opt['val'].get('grids', False):
          model.grids_inverse()

      visuals = model.get_current_visuals()
      sr_img = tensor2img([visuals['result']])
      imwrite(sr_img, save_path)
```

```python
import cv2
import numpy as np

def resize_np_image_with_opencv(np_image, scale=0.5):
    # np.arrayë¥¼ OpenCV í˜•ì‹ìœ¼ë¡œ ë³€í™˜ (ì´ë¯¸ OpenCV í˜•ì‹ì´ë¯€ë¡œ ë°”ë¡œ ì‚¬ìš© ê°€ëŠ¥)
    # ì´ë¯¸ì§€ í¬ê¸° ì¶•ì†Œ
    new_size = (int(np_image.shape[1] * scale), int(np_image.shape[0] * scale))
    resized_image = cv2.resize(np_image, new_size, interpolation=cv2.INTER_AREA)

    # ë¦¬ì‚¬ì´ì¦ˆëœ ì´ë¯¸ì§€ë¥¼ ë‹¤ì‹œ numpy ë°°ì—´ë¡œ ë°˜í™˜
    return resized_image

# ì˜ˆì‹œ
# ê¸°ì¡´ np.array ì´ë¯¸ì§€ ì˜ˆì œ ìƒì„±
original_image = np.random.randint(0, 256, (500, 500, 3), dtype=np.uint8)  # 500x500 ëœë¤ RGB ì´ë¯¸ì§€
resized_image = resize_np_image_with_opencv(original_image, scale=0.5)

print("Original Shape:", original_image.shape)
print("Resized Shape:", resized_image.shape)
```
